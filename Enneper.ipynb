```python
import numpy as np
import numba
import matplotlib as plt

import matplotlib.pyplot as plt
```

```python
def py_julia_fractal(z_re, z_im, j):
    '''Crea el grafico del fractal de Julia.'''
    for m in range(len(z_re)):
        for n in range(len(z_im)):
            z = z_re[m] + 1j * z_im[n]
            for t in range(256):
                z = z ** 2 - 0.05 + 0.68j
                if np.abs(z) > 2.0:
                    j[m, n] = t
                    break
jit_julia_fractal = numba.jit(nopython=True)(py_julia_fractal)
```

```python

N = 1024
j = np.zeros((N, N), np.int64)
z_real = np.linspace(-1.5, 1.5, N)
z_imag = np.linspace(-1.5, 1.5, N)
jit_julia_fractal(z_real, z_imag, j)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(j, cmap=plt.cm.RdBu_r, extent=[-1.5, 1.5, -1.5, 1.5])
ax.set_xlabel("$\mathrm{Re}(z)$", fontsize=18)
ax.set_ylabel("$\mathrm{Im}(z)$", fontsize=18)
plt.show()
```

# Expresion analitica para la formula de representación de Enneper




$\frac{1}{2}(1-\nu^2(z))μ(z))$

```python
A_1=1/2*(1-A*A)*B
C=simplify(A_1)
C
```

$\frac{i}{2}(1+ν^2(z))μ(z)$

```python
D=i/2*(1+A*A)*B
D
```

 $μ(z) ν(z)$

```python
E=A*B
E
```

$∫\frac{1}{2}(1-\nu^2(z))μ(z))dz$

```python
F=integrate(C,z)
F
```

$∫\frac{i}{2}(1+\nu^2(z))μ(z))dz$





```python
G=integrate(D,z)
G
```

$∫\nu(z)μ(z))dz$

```python
H=integrate(E,z)
H=simplify(H)
H
```

Por  lo tanto la superficie minima asociada a la formula de representacion de Enneper esta caracterizada por la parte real de la siguiente exprersión 

```python
J=[F,G,H]
J
```

\begin{equation*} 
\begin{split}

\end{split}
\end{equation*}

# Propiedades geometricas 

# Curvatura de Gauss

Como
 \begin{equation*}
\begin{split} 
 K(w)= -\left [  \frac{4⋅\left | v '(w)\right |}{\left | \mu(w) \right |(1+\left | v(w) \right |^2)^2} \right ]^2
\end{split}
\end{equation*}
Entonces vamos a tener que 

 \begin{equation*}
\begin{split} 
 K(w)&= -\left [  \frac{4⋅\left | 0.5(c-k)⋅e^{b+0.5(c-k)z}\right |}{\left | e^{a+k⋅z} \right |(1+\left | e^{b+0.5(c-k)z} \right |^2)^2} \right ]^2 \\
 &=-\left [  \frac{2(c-k)⋅ e^{a_2+0.5(c-k)u}}{ e^{a_1+k⋅u}⋅ (1+ e^{2⋅ a_2+(c-k)u})^2} \right ]^2 \\
 &=-\frac{4(c-k)^2 \cdot e^{2\cdot(a_1-a_2)}⋅e^{(c-3k)\cdot u}}{(1+ e^{2⋅ a_2}⋅e^{(c-k)u})^4}
\end{split}
\end{equation*}

# Ecuación de las lineas de curvatura 

Como todas las lineas de curvatura de una superficie que se pueda representar por medio de la formula de weiertrass estan caracterizadas por la ecuación 


$$ \Re e \left [ \mu (\gamma ) \cdot \dot{\nu }(\gamma ) \cdot \dot{\gamma }^2 \right ] = 0$$

Entonces como:

\begin{equation*}
\begin{split} 
\Re e \left [ \mu (\gamma ) \cdot \dot{\nu }(\gamma ) \cdot \dot{\gamma }^2 \right ]
&=\Re e\left [ e^{a+b+0.5\cdot (c+k)\cdot z(t)}\cdot( \dot{u}(t) +i\dot{v}(t))\right ]\\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\Re e\left [ e^{i(b_1+b_2)}e^{0.5 i\cdot (c+k)\cdot v(t)} (\dot{u}(t) +i\dot{v}(t))\right ]\\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\Re e\left [(\cos(b_1+b_2+0.5⋅(c+k)u(t))+i\sin( b_1+b_2+0.5⋅(c+k)v(t)))⋅(\dot{u}(t) +i\dot{v}(t))\right ] \\
&= e^{a_1+a_2}e^{0.5\cdot (c+k)\cdot u(t)}\left ( \cos(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)-\sin( b_1+b_2+0.5⋅(c+k)v(t)))⋅\dot{v}(t) \right )
\end{split}
\end{equation*}

Vamos a tener que las lineas de curvatura van a estar dados por la ecuación 

\begin{equation*}
\begin{split} 
 \cos(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)-\sin( b_1+b_2+0.5⋅(c+k)v(t))⋅\dot{v}(t)=0\\
\end{split}
\end{equation*}

# Ecuación de las lineas asintoticas 


Como todas las lineas asintoticas  de una superficie que se pueda representar por medio de la formula de weiertrass estan caracterizadas por la ecuación 

$$Im \left [ \mu (\gamma ) \cdot v '(\gamma ) \cdot \dot{\gamma }^2 \right ] = 0$$

Lo cual es equivalente a resolver 

\begin{equation*}
\begin{split}
\sin(b_1+b_2+0.5⋅(c+k)u(t))⋅ \dot{u}(t)+\cos( b_1+b_2+0.5⋅(c+k)v(t))⋅\dot{v}(t)=0
\end{split}
\end{equation*}

# Curvatura geodesica 



\begin{equation*}
\begin{split}
k_{g}&=\frac{1}{8}\left [\frac{\partial }{\partial u}\left ( \left | \mu(w) \right |^2(1+\left | v(w) \right |^2)^2 \right )\cdot \frac{dv}{ds} -\frac{\partial }{\partial v}\left ( \left | \mu(w) \right |^2(1+\left | v(w) \right |^2)^2 \right )\cdot  \frac{du}{ds}+4\cdot\frac{d}{ds} \arctan\left (\frac{\dot{u}(s))}{\dot{v}(s)}  \right )  \right ]\\

&=\left ( -\frac{1}{8} ke^{2cx-ku+a_1+4a_2}+\frac{1}{4}c
e^{2cu-ku+a_1+4a_2}+\frac{1}{4}ce^{cu+a_1+2a_2}+ke^{ku+a_1} \right )\frac{dv}{ds} +\frac{1}{2}\frac{d}{ds} \arctan\left (\frac{\dot{u}(s))}{\dot{v}(s)} \right )

\end{equation*}



# Ecuacion de las geodesicas 



\begin{equation*}
 \begin{split}
\ddot{u}+\frac{E_u}{2\cdot E}\cdot \dot{u} ^2+2\frac{E_v}{2\cdot E}\cdot \dot{u}  \dot{v} +\frac{G_u}{2\cdot E}\cdot \dot{v}^2=0\\

\end{equation*}

\end{split}
